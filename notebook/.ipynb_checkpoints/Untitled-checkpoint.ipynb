```python
class Image:
    def __init__(self, px, name=None):
        self._px = px
        self._tensor = torch.from_numpy((self._px.astype(np.float32)/255).transpose(2, 0, 1))
        self.name = name
    
    @property
    def px(self):
        return self._px
    
    @property
    def tensor(self):
        return self._tensor
```

```python
def open_rgby(path): #a function that reads RGBY image
    colors = ['red','green','blue','yellow']
    flags = cv2.IMREAD_GRAYSCALE
    img_id = path.name[:36]
    img = [cv2.imread(f"{str(path)}_{color}.png", flags).astype(np.uint8) for color in colors]
    return Image(px=np.stack(img, axis=-1), name=img_id)
```

```python
train_labels_one_hot = []
for labels in tqdm_notebook(train_labels):
    one_hot = np.zeros((28))
    for label in labels:
        one_hot[label] = 1
    train_labels_one_hot.append(one_hot.astype(np.float32))
```

```python
class ProteinClassificationDataset(torch.utils.data.Dataset):
    @staticmethod
    def collate_fn(batch):
        default_collate_batch = torch.utils.data.dataloader.default_collate(batch)
        x = default_collate_batch['input']
        label = default_collate_batch['label']
#         torch.from_numpy([e['label'] for e in batch])
        y = {
            'name': [e['name'] for e in batch],
            'label': label
        }
        return x, y
    
    def __init__(self, inputs, transform_fn=None, normalization_fn=None, labels=None):
        self.inputs = inputs
        self.transform_fn = transform_fn
        self.normalization_fn = normalization_fn
        self.labels = labels
        
    
    def __len__(self):
        return len(self.inputs)
    
    def __getitem__(self, i):
        ret = {}
        img = open_rgby(self.inputs[i])
        if self.transform_fn is not None:
            img = train_transform_fn(img)
        img_tensor = img.tensor
        if self.normalization_fn is not None:
            img_tensor = self.normalization_fn(img_tensor)
        ret['input'] = img_tensor
        ret['name'] = img.name
        if self.labels is not None:
            ret['label'] = self.labels[i]
        return ret
```

#### Build the the model

```python
import torch
import torch.nn as nn
import torchvision
```

```python
class Flatten(nn.Module):
    def __init__(self):
        super().__init__()
    def forward(self, x):
        return x.view(x.size(0), -1)

class AdaptiveConcatPool2d(nn.Module):
    def __init__(self):
        super().__init__()
        self.avg_pool = nn.AdaptiveAvgPool2d(1)
        self.max_pool = nn.AdaptiveMaxPool2d(1)
    
    def forward(self, x):
        return torch.cat([self.avg_pool(x), self.max_pool(x)], dim=1)

class Resnet50FourChannelInputConv(nn.Module):
    def __init__(self):
        super().__init__()
        pretrained_conv_weights = list(torchvision.models.resnet.resnet50(pretrained=True).children())[0].weight
        self.conv = nn.Conv2d(4, 64, kernel_size=7, stride=3, padding=3, bias=False)
        self.conv.weight = torch.nn.Parameter(torch.cat((pretrained_conv_weights, pretrained_conv_weights[:, :1, :, :]), dim=1))
    
    def forward(self, x):
        return self.conv(x)
```

#### Define the loss to be used

```python
class LossWrapper(nn.Module):
    def __init__(self, losses):
        super().__init__()
        self.losses = losses
        
    def forward(self, out, *yb):
        total_loss = sum([l(out, *yb) for l in self.losses])
        return total_loss
```

```python
class FocalLoss:
    def __init__(self, gamma=2):
        self.gamma = gamma
        
    def __call__(self, out, *yb):
        prediction = out
        target = yb[0]
        focal_loss = calculate_focal_loss(prediction, target, self.gamma)
        self.loss = focal_loss
        return focal_loss
```

```python
# https://becominghuman.ai/investigating-focal-and-dice-loss-for-the-kaggle-2018-data-science-bowl-65fb9af4f36c
import torch.nn.functional as F
def calculate_focal_loss(input, target, gamma=2):
    if not (target.size() == input.size()):
            raise ValueError("Target size ({}) must be the same as input size ({})"
                             .format(target.size(), input.size()))

    max_val = (-input).clamp(min=0)
    loss = input - input * target + max_val + \
        ((-max_val).exp() + (-input - max_val).exp()).log()

    invprobs = F.logsigmoid(-input * (target * 2.0 - 1.0))
    loss = (invprobs * gamma).exp() * loss

    return loss.sum(dim=1).mean()
```

```python
def acc(preds,targs,th=0.0):
    preds = (preds > th).int()
    targs = targs.int()
    return (preds==targs).float().mean()
```

#### Now do the training

```python
import pytorch_toolbox.fastai.fastai as fastai
from pytorch_toolbox.fastai_extensions.basic_train import Learner
```

```python
model = nn.Sequential(Resnet50FourChannelInputConv(),
                      *list(torchvision.models.resnet.resnet50(pretrained=True).children())[1:-2], 
                      AdaptiveConcatPool2d(),
                      Flatten(),
                      nn.BatchNorm1d(4096, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),
                      nn.Dropout(p=0.5),
                      nn.Linear(in_features=4096, out_features=2048, bias=True),
                      nn.BatchNorm1d(2048, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),
                      nn.Dropout(p=0.45),
                      nn.Linear(in_features=2048, out_features=1024, bias=True),
                      nn.BatchNorm1d(1024, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),
                      nn.Dropout(p=0.4),
                      nn.Linear(in_features=1024, out_features=512, bias=True),
                      nn.BatchNorm1d(512, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),
                      nn.Dropout(p=0.35),
                      nn.Linear(in_features=512, out_features=28, bias=True))
```

```python
n_starting_layers = len(fastai.flatten_model(model[:6]))
n_middle_layers = len(fastai.flatten_model(model[6:9]))
n_head = len(fastai.flatten_model(model[9:]))
layer_groups = fastai.split_model_idx(model, [n_starting_layers, n_starting_layers + n_middle_layers])
```

```python
class NameExtractionTrainer(fastai.Callback):
    def on_batch_begin(self, last_input, last_target, **kwargs):
        label = last_target.get('label')
        if label is not None:
            return last_input, last_target['label']
        else:
            return last_input, last_target
```

```python
learner = Learner(data,
                  layer_groups=layer_groups,
                  model=model, 
                  loss_func=LossWrapper([
                      FocalLoss()
                  ]),
                  callbacks=[NameExtractionTrainer()],
                  metrics=[acc])
```

```python
import warnings
warnings.filterwarnings('ignore')
learner.lr_find(start_lr=[1e-5] * 3, end_lr=[10] * 3, num_it=100)
```
